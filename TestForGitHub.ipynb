```python
# prompt: generate a graph that has a random varible that oscilates around x = y line on graph

import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(0, 10, 50)

# Generate random noise oscillating around x=y line
noise = np.random.normal(0, 1, 50)
y = x + noise

# Create the plot
plt.plot(x, y)
plt.plot(x, x, linestyle='--', color='red')  # Plot x=y line for reference

plt.xlabel('x')
plt.ylabel('y')
plt.title('Random Variable Oscillating Around x=y')

plt.show()
```
